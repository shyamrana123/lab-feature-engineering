```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsRegressor
from sklearn.neighbors import KNeighborsClassifier

from sklearn.preprocessing import MinMaxScaler, StandardScaler

from sklearn.metrics import precision_score, recall_score, classification_report, confusion_matrix, f1_score
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check for missing values**

```python
spaceship.isnull().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship.dropna(inplace=True)
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship["Cabin"]= spaceship["Cabin"].str[0]
spaceship["Cabin"].value_counts()
```

- Drop PassengerId and Name

```python
spaceship = spaceship.drop(columns = ["PassengerId", "Name"] )
```

- For non-numerical columns, do dummies.

```python
#your code here
spaceship = pd.get_dummies(spaceship, columns = ["HomePlanet", "Cabin", "Destination"], dtype= int)
```

```python
spaceship.columns
```

```python
spaceship[["CryoSleep","VIP"]]=spaceship[["CryoSleep","VIP"]].replace({"True":1 ,"False":0})
```

```python
spaceship.dtypes
```

**Perform Train Test Split**

```python
features = spaceship.drop(columns = ["Transported"])
target = spaceship["Transported"]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20)
```

```python
X_train
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
# Normalizing our data

normalizer = MinMaxScaler()
normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_test_norm = normalizer.transform(X_test)                                  # normalizing the test dataset as well
X_train_norm.head()
```

```python
#Training KNN to our normalized data

from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_norm,y_train)
```

- Evaluate your model's performance. Comment it

```python
#your code here
knn.predict(X_test_norm)
```

```python
y_test.values
```

```python
knn.score(X_test_norm,y_test)
```

```python
my_pred = knn.predict(X_test_norm)
```

```python
print(classification_report(y_pred= my_pred, y_true = y_test))
```
